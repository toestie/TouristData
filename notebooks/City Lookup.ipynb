```python
import pandas as pd
import requests
import json
from config import gkey

base_url = "https://maps.googleapis.com/maps/api/geocode/json"
params = {
    "key":gkey
}
```

```python
tourists_df = pd.read_csv("../data/GDCI_tourists.csv")
expenses_df = pd.read_csv("../data/GDCI_expenses.csv")
```

```python
cities = sorted(list(tourists_df["City"].append(expenses_df["City"]).unique()))
city_country_df = pd.DataFrame(cities,columns=["City"])
```

```python
for index, row in city_country_df.iterrows():
    params["address"] = row["City"]

    response = requests.get(base_url,params).json()
    country = response["results"][0]["address_components"][-1]["long_name"]
    latitude = response["results"][0]["geometry"]["location"]["lat"]
    longitude = response["results"][0]["geometry"]["location"]["lng"]
    
    city_country_df.loc[index,"Country"] = country
    city_country_df.loc[index,"City_Latitude"] = latitude
    city_country_df.loc[index,"City_Longitude"] = longitude
```

```python
# correct Pattaya country entry
city_country_df.loc[city_country_df["City"]=="Pattaya","Country"] = "Thailand"
city_country_df
```

```python
city_country_df.to_csv("../data/GDCI_city.csv",index=False)
```
