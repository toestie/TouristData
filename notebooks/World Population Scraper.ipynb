```python
import pandas as pd
from bs4 import BeautifulSoup
from splinter import Browser
```

```python
def init_browser():
    executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
    return Browser("chrome", **executable_path, headless=False)
```

```python
url = "http://worldpopulationreview.com/world-cities/amsterdam-population/"
```

```python
browser = init_browser()
browser.visit(url)
```

```python
soup = BeautifulSoup(browser.html, 'html.parser')
soup_table = soup.find_all("table")[0]
html_string = str(soup_table)
```

```python
amsterdam_pop_df = pd.read_html(html_string)[0]
amsterdam_pop_df
```

```python
amsterdam_pop_df.to_csv("amsterdam_population.csv")
```

```python
browser.quit()
```
